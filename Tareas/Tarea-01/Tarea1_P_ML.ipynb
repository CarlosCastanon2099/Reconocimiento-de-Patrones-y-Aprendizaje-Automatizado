##Use el método Newton-Raphson para encontrar una $raíz(f(x) = 0)$ y especifiqué el valor inicial $x_0 =?$ usado, también especifiqué la derivada de:(Nota no usar el valor de la raíz en $x_0$)

##(a) Ejemplo $f(x) = sen(x)$ con valor inicial $x_0 = 3$ y derivada $f′(x) = cos(x)$ da :
##Aproximación de la solución de $sin(x) = 0$: $3.141592653589793$.
##(b) $f(x) = cos(x)$
##(c) $f(x) = xsen(2x) + 1$
##(d) $f(x) = x^2 + 5x + 6$
##(e) $f(x) = x + e^x$



```python
%%html
<marquee style='width: 50%; color: blue; font-size: 40px;'><b>Solución</b></marquee>
```

```python

# Definimos una función para el método de Newton-Raphson
def newton_raphson(f, df, x0, epsilon=1e-6, max_iter=100):
    x = x0
    for _ in range(max_iter):
        x_n = x - f(x) / df(x)
        if abs(x - x_n) < epsilon:
            return x_n
        x = x_n
    return x

# Definimos las funciones y sus derivadas
functions = [
    (np.sin, np.cos),  # f(x) = sin(x), f'(x) = cos(x)
    (np.cos, lambda x: -np.sin(x)),  # f(x) = cos(x), f'(x) = -sin(x)
    (lambda x: x * np.sin(2*x) + 1, lambda x: np.sin(2*x) + 2*x*np.cos(2*x)),  # f(x) = xsin(2x)+1, f'(x) = sin(2x) + 2xcos(2x)
    (lambda x: x**2 + 5*x + 6, lambda x: 2*x + 5),  # f(x) = x^2 + 5x + 6, f'(x) = 2x + 5
    (lambda x: x + np.exp(x), lambda x: 1 + np.exp(x))  # f(x) = x + e^x, f'(x) = 1 + e^x
]

# Valor iniciales: Se eligen los puntos inciales x_0 para los incisos c a e de acorde al analisis visual de las graficas de cada funcion,
# se eigieron puntos cercanos  a una de las raices visibles de cada función
x0 = [3,-5,-1.5, -4,0]



# Aseguramos que la lista de funciones y la lista de x0s tengan la misma longitud
assert len(functions) == len(x0)



# Calculamos y graficamos las raíces para cada función
for i, ((f, df), x0) in enumerate(zip(functions,x0)):
    raiz = newton_raphson(f, df, x0)

    print(f"\n\nFunción {i+1}\tRaíz obtenida con newton-ralph -> {raiz}\n\n")

    # Graficamos la función y la raíz encontrada
    x = np.linspace(-10, 10, 400)
    y = f(x)

    plt.figure()
    plt.plot(x, y)
    plt.plot(raiz, f(raiz), 'ro')
    plt.axhline(0, color='black')  # Agregamos el eje x
    plt.title(f"Función {i+1} y su raíz")
    plt.grid(True)
    plt.show()


```

##(a) (0, 2),(1, −2)
##(b) (1, 2),(−1, −4)

```python
%%html
<marquee style='width: 50%; color: blue; font-size: 40px;'><b>Solución</b></marquee>
```

```python
def graficar_la_recta(puntos):
    # Crear una figura y un eje
    fig, ax = plt.subplots()

    # Un bucle para cada par de punto de a a d
    for i, ((x1, y1), (x2, y2)) in enumerate(puntos):
        # Calcular la pendiente y la intersección con el eje y usando la ecuacion
        m = (y2 - y1) / (x2 - x1)
        b = y1 - m * x1
        # imprimimos la pendiente de la recta en la terminal
        print(f'La pendiente de la recta {i+1} es: {m}')
        x = np.linspace(-10, 10, 400)
        y = m * x + b

        # Graficamos la recta
        ax.plot(x, y, label=f'Recta {i+1}')
        ax.plot([x1, x2], [y1, y2], 'ro')

    # datos de la grafica
    ax.legend()
    ax.grid(True)
    ax.axhline(0, color='black',linewidth=0.5)
    ax.axvline(0, color='black',linewidth=0.5)
    plt.title('Rectas que pasan por los puntos')
    plt.xlim([-10, 10])
    plt.ylim([-10, 10])
    plt.show()

# Puntos dados para cada inciso
a = [(0, 2), (1, -2)]
b = [(1, 2), (-1, -4)]
c = [(2, 2), (1, -6)]
d = [(3, 2), (-1, -8)]
puntos = [a, b,c,d]

# graficamos con la lista de puntos
graficar_la_recta(puntos)
```

### Genere una Interpolación Lineal usando n puntos de la función $cos(x)$ y $tan(x)$. Sobre el intervalo $[0, 2π]$. Indica cuantos puntos $n$ usaste. (Procura optimizar tu resultado).

```python
%%html
<marquee style='width: 50%; color: blue; font-size: 40px;'><b>Solución</b></marquee>
```

```python
# Función de interpolación lineal
def interpolar_lineal(x, y, x_interpolados):
    """
    Realiza una interpolación lineal para los puntos dados.

    Parameters:
        x (array): Arreglo de puntos x conocidos.
        y (array): Arreglo de puntos y conocidos correspondientes a x.
        x_interpolados (array): Arreglo de puntos x para la interpolación.

    Returns:
        array: Arreglo de puntos y interpolados correspondientes a x_interpolados.
    """
    y_interpolados = np.zeros_like(x_interpolados)
    for i in range(len(x_interpolados)):
        if x_interpolados[i] <= x[0]:
            y_interpolados[i] = y[0]
        elif x_interpolados[i] >= x[-1]:
            y_interpolados[i] = y[-1]
        else:
            j = np.searchsorted(x, x_interpolados[i])
            x0, x1 = x[j-1], x[j]
            y0, y1 = y[j-1], y[j]
            y_interpolados[i] = y0 + ((y1 - y0) / (x1 - x0)) * (x_interpolados[i] - x0)
    return y_interpolados

# Función para graficar la interpolación
def graficar_interpolacion(x_valores, y_valores, nombres_funciones):
    """
    Grafica la interpolación lineal para varios conjuntos de datos.

    Parameters:
        x_valores (list): Lista de arreglos de puntos x.
        y_valores (list): Lista de arreglos de puntos y correspondientes a x.
        nombres_funciones (list): Lista de nombres de funciones para mostrar en la leyenda.
    """
    fig, axs = plt.subplots(1, len(x_valores), figsize=(15, 8))

    for i in range(len(x_valores)):
        # Puntos para la interpolación
        x_interpolados = np.linspace(x_valores[i][0], x_valores[i][-1], 1000)
        y_interpolados = interpolar_lineal(x_valores[i], y_valores[i], x_interpolados)

        axs[i].plot(x_interpolados, y_interpolados, label=f'Interpolación lineal de {nombres_funciones[i]}(x)')
        axs[i].scatter(x_valores[i], y_valores[i], color='red', marker='o', label=f'Puntos de {nombres_funciones[i]}(x)')
        axs[i].set_xlabel('x')
        axs[i].set_ylabel(f'{nombres_funciones[i]}(x)')
        axs[i].set_title(f'Interpolación lineal de {nombres_funciones[i]}(x) con {len(x_valores[i])} puntos',fontsize=9)
        axs[i].legend(fontsize=8)

    plt.tight_layout()
    plt.show()


# Lista de valores de n para la interpolación
n_valores = [5, 10, 20, 50]

# Generar puntos equidistantes en el intervalo [0, 2π] para cada valor de n
x_valores_coseno = [np.linspace(0, 2 * np.pi, n) for n in n_valores]
x_valores_tangente = [np.linspace(0, 2 * np.pi, n) for n in n_valores]

# Calcular la función coseno y tangente para cada conjunto de puntos x
y_valores_coseno = [np.cos(x) for x in x_valores_coseno]
y_valores_tangente = [np.tan(x) for x in x_valores_tangente]
```

# 6.

##Genere una Interpolación Polinómica de Newton usando n puntos de la función $cos(x)$ y $tan(x)$. Sobre el intervalo $[0, 2π]$. Indica cuantos puntos n usaste.(Procura optimizar tu resultado).

```python
%%html
<marquee style='width: 50%; color: blue; font-size: 40px;'><b>Solución</b></marquee>
```

```python
# Definir la función cos(x) y tan(x)
def f_cos(x):
    return np.cos(x)

def f_tan(x):
    return np.tan(x)

# Definir la función para calcular los puntos de interpolación
def calcular_puntos_interpolacion(funcion, n):
    x_interpolacion = np.linspace(0, 2 * np.pi, n)
    y_interpolacion = funcion(x_interpolacion)
    return x_interpolacion, y_interpolacion

# Definir la función para calcular el polinomio de Newton
def polinomio_newton(x, y):
    n = len(x)
    coeficientes = y.copy()
    for j in range(1, n):
        coeficientes[j:] = (coeficientes[j:] - coeficientes[j - 1]) / (x[j:] - x[j - 1])
    def polinomio(x_eval):
        resultado = coeficientes[-1]
        for j in range(n - 2, -1, -1):
            resultado = resultado * (x_eval - x[j]) + coeficientes[j]
        return resultado
    return polinomio

# Número de puntos de interpolación
n_valores = [5, 7, 20, 50]

# Calcular puntos de interpolación y graficar para cada valor de n
for n in n_valores:
    x_cos, y_cos = calcular_puntos_interpolacion(f_cos, n)
    x_tan, y_tan = calcular_puntos_interpolacion(f_tan, n)

    polinomio_cos = polinomio_newton(x_cos, y_cos)
    polinomio_tan = polinomio_newton(x_tan, y_tan)

    x_eval = np.linspace(0, 2 * np.pi, 100)
    y_eval_cos = polinomio_cos(x_eval)
    y_eval_tan = polinomio_tan(x_eval)

    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.title(f'Interpolación de cos(x) con {n} puntos')
    plt.plot(x_eval, y_eval_cos, label='Interpolación')
    plt.plot(x_eval, f_cos(x_eval), label='cos(x)')
    plt.scatter(x_cos, y_cos, color='red', label='Puntos de interpolación')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.title(f'Interpolación de tan(x) con {n} puntos')
    plt.plot(x_eval, y_eval_tan, label='Interpolación')
    plt.plot(x_eval, f_tan(x_eval), label='tan(x)')
    plt.scatter(x_tan, y_tan, color='red', label='Puntos de interpolación')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

### Genere una Interpolación Polinómica de Lagrange usando n puntos de la función $cos(x)$ y $tan(x)$. Sobre el intervalo $[0, 2π]$. Indica cuantos puntos $n$ usaste y el polinomio obtenido. (Procura optimizar tu resultado).

```python
# Funcion de Interpolacion Polinomica de Lagrange
def interpolar_lagrange(x, y, x_interpolados):
    """
    Realiza una interpolación polinómica de Lagrange para los puntos dados.

    Parameters:
        x (array): Arreglo de puntos x conocidos.
        y (array): Arreglo de puntos y conocidos correspondientes a x.
        x_interpolados (array): Arreglo de puntos x para la interpolación.

    Returns:
        array: Arreglo de puntos y interpolados correspondientes a x_interpolados.
    """
    y_interpolados = np.zeros_like(x_interpolados)
    for i in range(len(x_interpolados)):
        for j in range(len(x)):
            numerador = 1
            denominador = 1
            for k in range(len(x)):
                if k != j:
                    numerador *= (x_interpolados[i] - x[k])
                    denominador *= (x[j] - x[k])
            y_interpolados[i] += y[j] * (numerador / denominador)
    return y_interpolados

# Función para graficar la interpolación de Lagrange
def graficar_interpolacion_lagrange(x_valores, y_valores, nombres_funciones):
    """
    Grafica la interpolación polinómica de Lagrange para varios conjuntos de datos.

    Parameters:
        x_valores (list): Lista de arreglos de puntos x.
        y_valores (list): Lista de arreglos de puntos y correspondientes a x.
        nombres_funciones (list): Lista de nombres de funciones para mostrar en la leyenda.
    """
    fig, axs = plt.subplots(1, len(x_valores), figsize=(15, 8))

    for i in range(len(x_valores)):
        # Puntos para la interpolación
        x_interpolados = np.linspace(x_valores[i][0], x_valores[i][-1], 1000)
        y_interpolados = interpolar_lagrange(x_valores[i], y_valores[i], x_interpolados)

        axs[i].plot(x_interpolados, y_interpolados, label=f'Interpolación polinómica de Lagrange de {nombres_funciones[i]}(x)')
        axs[i].scatter(x_valores[i], y_valores[i], color='red', marker='o', label=f'Puntos de {nombres_funciones[i]}(x)')
        axs[i].set_xlabel('x')
        axs[i].set_ylabel(f'{nombres_funciones[i]}(x)')
        axs[i].set_title(f'Interpolación polinómica de Lagrange de {nombres_funciones[i]}(x) con {len(x_valores[i])} puntos',fontsize=8)
        axs[i].legend(fontsize=6.5)

    plt.tight_layout()
    plt.show()

# Lista de valores de n para la interpolación
n_valoresLagrange = [5, 10, 20, 50]

# Generar puntos equidistantes en el intervalo [0, 2π] para cada valor de n
x_valores_cosenoLagrange = [np.linspace(0, 2 * np.pi, n) for n in n_valoresLagrange]
x_valores_tangenteLagrange = [np.linspace(0, 2 * np.pi, n) for n in n_valoresLagrange]

# Calcular la función coseno y tangente para cada conjunto de puntos x
y_valores_cosenoLagrange = [np.cos(x) for x in x_valores_cosenoLagrange]
y_valores_tangenteLagrange = [np.tan(x) for x in x_valores_tangenteLagrange]
```

```python
graficar_interpolacion_lagrange(x_valores_tangenteLagrange, y_valores_tangenteLagrange, "tangente")
```

```python
# Definir las funciones  coseno (cos(x)) y tangente (tan(x))
def f_cos(x):
    return np.cos(x)

def f_tan(x):
    return np.tan(x)

# Números de puntos de interpolación
n_valores = [4, 5, 20, 50]

# Grado de la B-spline (puedes experimentar con diferentes valores)
k = 2

# Intervalo [0, 2π]
x_intervalo = np.linspace(0, 2 * np.pi, 100)

# Graficar para cada valor de n
for n in n_valores:
    # Generar puntos de interpolación para cos(x) y tan(x)

    x_cos = np.linspace(0, 2 * np.pi, n)
    y_cos = f_cos(x_cos)
    x_tan = np.linspace(0, 2 * np.pi, n)
    y_tan = f_tan(x_tan)

    # Calcular B-spline para cos(x)
    tck_cos = splrep(x_cos, y_cos, k=k)

    # Calcular B-spline para tan(x)
    tck_tan = splrep(x_tan, y_tan, k=k)

    # Evaluar los B-splines en el intervalo
    spline_cos = splev(x_intervalo, tck_cos)
    spline_tan = splev(x_intervalo, tck_tan)

    # Graficar cos(x) y tan(x) junto con sus B-splines
    plt.figure(figsize=(18, 4))

    plt.subplot(1, 3, 2)
    plt.title(f'B-spline de cos(x) con {n} puntos (k={k})')
    plt.plot(x_intervalo, spline_cos, label='B-spline')
    plt.plot(x_cos, y_cos, 'o', label='Valores conocidos')
    plt.plot(x_intervalo, f_cos(x_intervalo), '--', label='cos(x)')
    plt.legend()

    plt.subplot(1, 3, 3)
    plt.title(f'B-spline de tan(x) con {n} puntos (k={k})')
    plt.plot(x_intervalo, spline_tan, label='B-spline')
    plt.plot(x_tan, y_tan, 'o', label='Valores conocidos')
    plt.plot(x_intervalo, f_tan(x_intervalo), '--', label='tan(x)')
    plt.legend()

    plt.tight_layout()
    plt.show()

```
